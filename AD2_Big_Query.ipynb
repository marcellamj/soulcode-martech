**Verificar registros por região**
```
# Contagem por regiao
SELECT _regiao_, count(*) as qtd
FROM projetoooo-390016.dados_covid.dados_covid_brazil
GROUP BY _regiao_;

# Retirar a inconsistência _regiao_ e as aspas usando replace ou rename
SELECT REPLACE (_regiao_, "'","") AS regiao,
COUNT(_regiao_) as qtd
FROM `projetoooo-390016.dados_covid.dados_covid_brazil`
group by _regiao_;
```

**Verificar registros por estado**

```
# Contagem por estado
SELECT _estado_, count(*) as qtd
FROM projetoooo-390016.dados_covid.dados_covid_brazil
GROUP BY _estado_;

# Retirar a inconsistência _estado_ e as aspas usando replace ou rename
# Somar novos obitos como qtd
# Excluir dados nulos do filtro
# Agrupar por estado e mostrar os 10 com mais obitos

SELECT REPLACE (_estado_, "'","") AS municipio,
SUM(_obitosnovos_) as obitos
FROM `projetoooo-390016.dados_covid.dados_covid_brazil`
WHERE _estado_ IS NOT NULL
GROUP BY _estado_
ORDER BY qtd DESC
LIMIT 5;
```
